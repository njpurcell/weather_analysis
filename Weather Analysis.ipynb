```python
import re
```

```python
import numpy as np
```

```python
def get_temp(x):
    if re.match(r'\d\d', x[:2]):
        return int(x[:2])
    else:
        search = re.findall(r'\d\d', x)
        if search:
            if int(search[0]) > 22:
                return int(search[0])
            elif len(search) > 1 and int(search[1]) > 22:
                return int(search[1])
    return np.nan
```

```python
data_no_na["temperature"] = data_no_na["weather"].apply(get_temp)
```

```python
data_no_na.head()
```

```python
data_no_na["runs_scored"] = data_no_na["visitor_final_score"].astype(int) + data_no_na["home_final_score"].astype(int)
```

```python
data_no_na = data_no_na[pd.notnull(data_no_na["temperature"])]
data_no_na.shape[0]
```

```python
from sklearn.utils import shuffle
```

```python
data_no_na = shuffle(data_no_na)
x = data_no_na["temperature"]
y = data_no_na["runs_scored"]
#lists = sorted(zip(*[x, y]))
#new_x, new_y = list(zip(*lists))
#new_x = list(new_x)
#new_y = list(new_y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x=x.reshape(len(x),1)
y=y.reshape(len(y),1)
 
# Split the data into training/testing sets
x_train = x[:-120]
x_test = x[-70:]
 
# Split the targets into training/testing sets
y_train = y[:-120]
y_test = y[-120:]
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
predictions = reg.predict(x_test)
```

```python
plt.scatter(x,y, color="#00bbff")
plt.title("IIAC Temperature v. Runs Scored", fontsize=17)
plt.ylabel("Runs Scored")
plt.xlabel("Temperature (Fahrenheit)")
plt.plot(x_test, predictions, color='red',linewidth=3)
plt.show()
```

```python
# R^2
reg.score(x,y)
```

```python
reg.coef_
```

There isn't much here. If you squint real hard at the scatter plot (or look at the coefficient), you'll find that for every 10 degree increase in temperature, we can expect one additional run to be scored per game.
