```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data17 = pd.read_csv("games2017.csv")
data17.head()
```

```python
data18 = pd.read_csv("games2018.csv")
data18.head()
```

```python
data = pd.concat([data17,data18])
```

```python
data.columns
```

```python
data["weather"].head(20)
```

```python
data_no_na = data[pd.notnull(data["weather"])]
data_no_na.shape[0]
```

```python
data_no_na["weather"].head(25)
```

```python
for i in range(data_no_na.shape[0]):
    print(data_no_na.iloc[i]["weather"])
```

```python
import re
```

```python
import numpy as np
```

```python
def get_temp(x):
    if re.match(r'\d\d', x[:2]):
        return int(x[:2])
    else:
        search = re.findall(r'\d\d', x)
        if search:
            if int(search[0]) > 22:
                return int(search[0])
            elif len(search) > 1 and int(search[1]) > 22:
                return int(search[1])
    return np.nan
```

```python
data_no_na["temperature"] = data_no_na["weather"].apply(get_temp)
```

```python
data_no_na.head()
```

```python
data_no_na["runs_scored"] = data_no_na["visitor_final_score"].astype(int) + data_no_na["home_final_score"].astype(int)
```

```python
data_no_na = data_no_na[pd.notnull(data_no_na["temperature"])]
data_no_na.shape[0]
```

```python
from sklearn.utils import shuffle
```

```python
data_no_na = shuffle(data_no_na)
X = data_no_na[["temperature"]]
y = data_no_na["runs_scored"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
predictions = reg.predict(X)
```

```python
plt.scatter(X,y, color="#00bbff")
plt.title("IIAC Temperature v. Runs Scored", fontsize=17)
plt.ylabel("Runs Scored")
plt.xlabel("Temperature (Fahrenheit)")
plt.plot(X, predictions, color='red',linewidth=3)
plt.show()
```

```python
# R^2
reg.score(X,y)
```

```python
reg.coef_
```

There isn't much here. If you squint real hard at the scatter plot (or look at the coefficient), you'll find that for every 12.5 degree increase in temperature, we can expect one additional run to be scored per game.
